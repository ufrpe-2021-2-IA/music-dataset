```python
import pandas as pd
```

# Min-Max

```python
df = pd.read_csv('../gtzan/processed/mgd_raw.csv')
df.describe()
```

```python
features_df = df.iloc[:, 2:]
features_df = (features_df - features_df.min()) / (features_df.max() - features_df.min())
df.iloc[:, 2:] = features_df
df.to_csv('../gtzan/processed/mgd_min_max.csv', index=False)
```

# Standardization

```python
df = pd.read_csv('../gtzan/processed/mgd_raw.csv')
df.describe()
```

```python
features_df = df.iloc[:, 2:]
features_df = (features_df - features_df.mean()) / features_df.std()
df.iloc[:, 2:] = features_df
df.to_csv('../gtzan/processed/mgd_standardized.csv', index=False)
```
